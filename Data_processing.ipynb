```python
def haversine_distance(lat1, lon1, lat2, lon2):
    # Bán kính Trái đất theo km
    r = 6371 
    
    # Chuyển đổi sang radian
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    dphi = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    
    # Công thức Haversine
    a = np.sin(dphi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(dlambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    
    return r * c

# Giả sử df là DataFrame của bạn
# Bước 1: Tạo một bản lọc các dòng bị null ở cột khoảng cách
mask = df['Transportation Distance (KM)'].isnull()

# Bước 2: Áp dụng công thức (chỉ áp dụng cho những dòng bị thiếu)
df.loc[mask, 'Transportation Distance (KM)'] = df.apply(
    lambda row: haversine_distance(
        row['Origin Location Latitude'], 
        row['Origin Location Longitude'], 
        row['Destination Location Latitude'], 
        row['Destination Location Longitude']
    ) if pd.isnull(row['Transportation Distance (KM)']) else row['Transportation Distance (KM)'],
    axis=1
)

# Bước 3: Logistics thực tế thường đường bộ sẽ dài hơn đường chim bay
# Bạn có thể nhân với hệ số "Detour Factor" (thường là 1.2 đến 1.4 tùy địa hình)
# df.loc[mask, 'Transportation Distance (KM)'] *= 1.2

print(f"Đã xử lý xong. Số dòng còn trống: {df['Transportation Distance (KM)'].isnull().sum()}")
```

```python
def classify_vehicle(v_type):
    v_type = str(v_type).upper()
    
    if v_type == 'NAN' or v_type == 'NONE':
        return 'Unknown'
    elif 'TRAILER' in v_type or '40 FT' in v_type or 'FLAT BED' in v_type:
        return 'Heavy Trailer (40FT+)'
    elif 'CONTAINER' in v_type:
        return 'Container'
    elif 'HCV' in v_type or '32 FT' in v_type or 'TAURUS' in v_type or 'MXL' in v_type:
        return 'Heavy Commercial (HCV)'
    elif 'MCV' in v_type or '17 FT' in v_type or '19 FT' in v_type or '20 FT' in v_type:
        return 'Medium Commercial (MCV)'
    elif 'LCV' in v_type or 'TATA ACE' in v_type or 'PICKUP' in v_type or '1 MT' in v_type or '2MT' in v_type:
        return 'Light Commercial (LCV/Pickup)'
    else:
        return 'Other'

# Tạo cột mới Vehicle Category
df['Vehicle Category'] = df['Vehicle Type'].apply(classify_vehicle)

# Kiểm tra kết quả
print(df['Vehicle Category'].value_counts())
```

```python
import pandas as pd
import numpy as np
import os

# --- BƯỚC 0: ĐỌC DỮ LIỆU ---
file_input = 'Data_logistics_clean.csv'
try:
    df = pd.read_csv(file_input)
    print(f"Đã nạp thành công dữ liệu từ {file_input}")
except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file {file_input}")
    exit()

# Chuẩn hóa tên cột (Xóa khoảng trắng và ký tự đặc biệt)
df.columns = [c.replace(' ', '_').replace('(', '').replace(')', '').replace('/', '_') for c in df.columns]

# --- 1. Dim_Locations (Danh mục địa điểm duy nhất) ---
# Gom tất cả tọa độ và tên địa điểm vào một bảng Master
loc_origin = df[['Origin_Location', 'Origin_loc_latitude', 'Origin_loc_longitude']].rename(
    columns={'Origin_Location': 'Location_Name', 'Origin_loc_latitude': 'Latitude', 'Origin_loc_longitude': 'Longitude'})
loc_dest = df[['Destination_Location', 'Destination_loc_latitude', 'Destination_loc_longitude']].rename(
    columns={'Destination_Location': 'Location_Name', 'Destination_loc_latitude': 'Latitude', 'Destination_loc_longitude': 'Longitude'})

dim_locations = pd.concat([loc_origin, loc_dest]).drop_duplicates(subset=['Location_Name']).reset_index(drop=True)
dim_locations['Location_Key'] = dim_locations.index + 1

# --- 2. Dim_Vehicle_Driver ---
dim_vehicle_driver = df[['Vehicle_Registration', 'Vehicle_Type', 'Vehicle_Category', 'Driver_Name']].drop_duplicates().reset_index(drop=True)
dim_vehicle_driver['Vehical_Driver_Key'] = dim_vehicle_driver.index + 1

# --- 3. Dim_Partners_Materials ---
dim_partners = df[['Customer_Name', 'Supplier_Name', 'Material_Shipped', 'Shipment_Type', 'Gps_Provider']].drop_duplicates().reset_index(drop=True)
dim_partners['Partner_Material_Key'] = dim_partners.index + 1

# --- 4. Dim_Date ---
# Đảm bảo Booking_Date là datetime
df['Booking_Date'] = pd.to_datetime(df['Booking_Date'])
dim_date = pd.DataFrame({'Full_Date': pd.to_datetime(df['Booking_Date'].unique())})
dim_date['Date_Key'] = dim_date['Full_Date'].dt.strftime('%Y%m%d').astype(int)
dim_date['Day_Of_Week'] = dim_date['Full_Date'].dt.day_name()
dim_date['Is_Weekend'] = dim_date['Full_Date'].dt.dayofweek.apply(lambda x: 1 if x >= 5 else 0)
dim_date['Month'] = dim_date['Full_Date'].dt.month
dim_date['Quarter'] = dim_date['Full_Date'].dt.quarter
dim_date['Year'] = dim_date['Full_Date'].dt.year

# --- 5. Fact_Shipments (Bảng sự kiện chính) ---
fact = df.copy()

# Chuyển đổi các cột thời gian sang định dạng chuẩn
time_cols = ['trip_start_date', 'trip_end_date', 'Actual_ETA', 'Planned_ETA']
for col in time_cols:
    if col in fact.columns:
        fact[col] = pd.to_datetime(fact[col])

# Map các Key từ bảng Dim vào Fact
fact = fact.merge(dim_vehicle_driver, on=['Vehicle_Registration', 'Vehicle_Type', 'Vehicle_Category', 'Driver_Name'], how='left')
fact = fact.merge(dim_partners, on=['Customer_Name', 'Supplier_Name', 'Material_Shipped', 'Shipment_Type', 'Gps_Provider'], how='left')

# Map Location Keys cho cả điểm đi và điểm đến
fact = fact.merge(dim_locations[['Location_Key', 'Location_Name']].rename(columns={'Location_Key': 'Location_Origin_Key', 'Location_Name': 'Origin_Location'}), on='Origin_Location', how='left')
fact = fact.merge(dim_locations[['Location_Key', 'Location_Name']].rename(columns={'Location_Key': 'Location_Dest_Key', 'Location_Name': 'Destination_Location'}), on='Destination_Location', how='left')

# Map Date Key
fact['Date_Key'] = fact['Booking_Date'].dt.strftime('%Y%m%d').astype(int)

# Tính toán thêm các cột Measure nếu cần (Ví dụ: Travel Duration tính bằng phút)
if 'trip_start_date' in fact.columns and 'trip_end_date' in fact.columns:
    fact['Travel_Duration'] = (fact['trip_end_date'] - fact['trip_start_date']).dt.total_seconds() / 60

# Chọn đúng danh sách cột bạn đã yêu cầu cho Fact
fact_final_columns = [
    'Booking_ID', 'Vehical_Driver_Key', 'Partner_Material_Key', 
    'Location_Origin_Key', 'Location_Dest_Key', 'Date_Key',
    'Transportation_Distance_KM', 'Travel_Duration', 'Actual_ETA', 
    'Planned_ETA', 'trip_start_date', 'trip_end_date', 'Ontime'
]

# Kiểm tra xem cột nào tồn tại thì mới lấy cột đó
fact_shipments = fact[[c for c in fact_final_columns if c in fact.columns]]

# --- BƯỚC CUỐI: XUẤT FILE EXCEL ĐA SHEET ---
output_file = 'Logistics_Star_Schema_Final.xlsx'
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    fact_shipments.to_excel(writer, sheet_name='Fact', index=False)
    dim_locations.to_excel(writer, sheet_name='Dim_Locations', index=False)
    dim_vehicle_driver.to_excel(writer, sheet_name='Dim_Vehicle_Driver', index=False)
    dim_partners.to_excel(writer, sheet_name='Dim_Partners_Materials', index=False)
    dim_date.to_excel(writer, sheet_name='Dim_Date', index=False)

print("-" * 30)
print(f"THÀNH CÔNG RỰC RỠ!")
print(f"File đã sẵn sàng tại: {os.path.abspath(output_file)}")
print("Giờ bạn chỉ cần nạp file này vào Looker Studio và bắt đầu kéo thả Dashboard.")
```
